```python
import pandas as pd
from faker import Faker
import random

# Initialize Faker for synthetic data generation
fake = Faker()

# Define the number of rows to generate
num_rows = 100

# Define which columns are numeric and which are categorical
numeric_columns = [
    'Age', 'Daily Screen Time','Smartphone Usage Time Slots',
    'Frequency of Checking the Phone  (number of times per hour)',
    'Unlock counts per hour', 'Notification density',
    'Total and Specific screen time after 10 pm (in hours)',
    'Video streaming time', 'Gaming engagement time',
    'Audio streaming time'
]

# Categorical columns to be populated with Faker data
categorical_columns = [
    'Name', 'Gender', 'Occupation/Profession', 'Education Level',
    'Most Used Apps', 'Response to notifications',
    'App specific screen time (Mention App - Screen Time)'
]

# Create an empty dictionary to hold synthetic data
synthetic_data = {}

# Generate synthetic data for categorical columns
for col in categorical_columns:
    if col == 'Name':
        synthetic_data[col] = [fake.name() for _ in range(num_rows)]
    elif col == 'Gender':
        synthetic_data[col] = [random.choice(['Male', 'Female', 'Other']) for _ in range(num_rows)]
    elif col == 'Occupation/Profession':
        synthetic_data[col] = [fake.job() for _ in range(num_rows)]
    elif col == 'Education Level':
        synthetic_data[col] = [random.choice(['High School', 'Undergraduate', 'Graduate', 'Postgraduate']) for _ in range(num_rows)]
    elif col == 'Most Used Apps':
        synthetic_data[col] = [fake.word() for _ in range(num_rows)]
    elif col == 'Response to notifications':
        synthetic_data[col] = [random.choice(['Immediate', 'Delayed', 'Not Responded']) for _ in range(num_rows)]
    elif col == 'App specific screen time (Mention App - Screen Time)':
        synthetic_data[col] = [fake.word() + " - " + str(random.randint(1, 6)) for _ in range(num_rows)]  # Lowered the screen time to 1-6 hours

# Generate synthetic data for numeric columns with more realistic values
for col in numeric_columns:
    if col == 'Age':
        synthetic_data[col] = [random.randint(18, 60) for _ in range(num_rows)]  # Age between 18 and 60
    elif col == 'Daily Screen Time':
        synthetic_data[col] = [random.randint(1, 20) for _ in range(num_rows)]
    elif col == 'Smartphone Usage Time Slots':
        synthetic_data[col] = [random.randint(1, 10) for _ in range(num_rows)]  # Usage slots between 1 and 10
    elif col == 'Frequency of Checking the Phone  (number of times per hour)':
        synthetic_data[col] = [random.randint(5, 20) for _ in range(num_rows)]  # Checking frequency between 5 and 20 times per hour
    elif col == 'Unlock counts per hour':
        synthetic_data[col] = [random.randint(5, 30) for _ in range(num_rows)]  # Unlock counts per hour between 5 and 30
    elif col == 'Notification density':
        synthetic_data[col] = [random.randint(1, 50) for _ in range(num_rows)]  # Notification density between 1 and 50
    elif col == 'Total and Specific screen time after 10 pm (in hours)':
        synthetic_data[col] = [round(random.uniform(0.5, 4), 2) for _ in range(num_rows)]  # Screen time after 10 pm between 0.5 and 4 hours
    elif col == 'Video streaming time':
        synthetic_data[col] = [round(random.uniform(0, 10), 2) for _ in range(num_rows)]  # Video streaming time between 0 and 10 hours
    elif col == 'Gaming engagement time':
        synthetic_data[col] = [round(random.uniform(0, 10), 2) for _ in range(num_rows)]  # Gaming engagement time between 0 and 10 hours
    elif col == 'Audio streaming time':
        synthetic_data[col] = [round(random.uniform(0, 10), 2) for _ in range(num_rows)]  # Audio streaming time between 0 and 10 hours

# Create a DataFrame from the synthetic data
df_synthetic = pd.DataFrame(synthetic_data)

# Save the synthetic data to a CSV file
df_synthetic.to_csv('synthetic_data_100_rows.csv', index=False)

# Display the first few rows of the synthetic data
print(df_synthetic.head())
print(df_synthetic.columns)
```

```python
import pandas as pd
import numpy as np
import joblib  # For saving and loading model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
df = pd.read_csv('merged_dataset.csv')

# Drop rows with missing values
df.dropna(inplace=True)

# Convert numerical columns to proper data types
numerical_features = ['Daily Screen Time', 'Unlock counts per hour', 'Video streaming time', 'Gaming engagement time']
df[numerical_features] = df[numerical_features].apply(pd.to_numeric, errors='coerce')

# Encode categorical variables
label_encoders = {}
categorical_features = ['Gender', 'Occupation/Profession', 'Education Level', 'Most Used Apps', 'Response to notifications']
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Convert or classify stress level
if 'Stress Level' in df.columns:
    stress_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
    df['Stress Level'] = df['Stress Level'].map(stress_mapping)
else:
    def classify_stress(row):
        if row['Daily Screen Time'] > 8 or row['Unlock counts per hour'] > 20 or row['Video streaming time'] > 5 or row['Gaming engagement time'] > 3:
            return 'High'
        elif row['Daily Screen Time'] > 5 or row['Unlock counts per hour'] > 10 or row['Video streaming time'] > 2 or row['Gaming engagement time'] > 1:
            return 'Medium'
        else:
            return 'Low'

    df['Stress Level'] = df.apply(classify_stress, axis=1)
    df['Stress Level'] = df['Stress Level'].map({'Low': 0, 'Medium': 1, 'High': 2})

# Drop remaining NaN values
df.dropna(inplace=True)

# Select key features and target variable
selected_features = ['Daily Screen Time', 'Unlock counts per hour', 'Video streaming time', 'Gaming engagement time']
X = df[selected_features]
y = df['Stress Level']

# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Save the trained model and scaler
joblib.dump(rf, 'stress_model.pkl')
joblib.dump(scaler, 'scaler.pkl')

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_test, y_pred))

# Feature importance
feature_importance = pd.DataFrame({'Feature': selected_features, 'Importance': rf.feature_importances_}).sort_values(by='Importance', ascending=False)
print(feature_importance)

# --------------------- PREDICTION ON USER INPUT ---------------------

# Load model and scaler
loaded_model = joblib.load('stress_model.pkl')
loaded_scaler = joblib.load('scaler.pkl')

# Get user input for stress prediction (only 4 features)
daily_screen_time = float(input("Enter your Daily Screen Time (in hours): "))
unlock_counts = float(input("Enter your Unlock counts per hour: "))
video_streaming_time = float(input("Enter your Video streaming time (in hours): "))
gaming_time = float(input("Enter your Gaming engagement time (in hours): "))

# Convert to numpy array
user_data = np.array([[daily_screen_time, unlock_counts, video_streaming_time, gaming_time]])

# Scale input data
user_data_scaled = loaded_scaler.transform(user_data)

# Predict stress level
predicted_stress = loaded_model.predict(user_data_scaled)[0]

# Convert numerical prediction back to stress level category
stress_labels = {0: 'Low', 1: 'Medium', 2: 'High'}
print(f'Predicted Stress Level: {stress_labels[predicted_stress]}')
```

```python
import google.generativeai as genai
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# ðŸ”¹ Set up Gemini Flash 1.5 API Key
genai.configure(api_key="API_KEY")

# ðŸ”¹ Initialize Gemini Flash Model
model_gemini = genai.GenerativeModel("gemini-1.5-flash")

# ðŸ”¹ Sample Training Data
X_train = np.array([[8, 5, 4, 3], [5, 3, 2, 1], [10, 7, 6, 5], [4, 2, 1, 0]])
y_train = np.array(["High", "Low", "High", "Low"])

# ðŸ”¹ Train Random Forest Model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train_scaled, y_train)

# ðŸ”¹ Get User Input
age=int(input("Enter your age: "))
occupation=input("Enter your occupation: ")
education=input("Enter your education level: ")
daily_screen_time = float(input("Enter your Daily Screen Time (hours): "))
unlock_counts = float(input("Enter Unlock counts per hour: "))
video_time = float(input("Enter Video streaming time (hours): "))
gaming_time = float(input("Enter Gaming engagement time (hours): "))

# ðŸ”¹ Scale Input & Predict Stress Level
user_data = np.array([[daily_screen_time, unlock_counts, video_time, gaming_time]])
user_data_scaled = scaler.transform(user_data)
predicted_stress = model_rf.predict(user_data_scaled)[0]

# ðŸ”¹ Generate AI Report Using Gemini Flash
prompt = f"""
A user has the following smartphone usage patterns:
-Age: {age}
-Occupation: {occupation}
-Education Level: {education}
- Daily Screen Time: {daily_screen_time} hours
- Unlocks per hour: {unlock_counts}
- Video Streaming Time: {video_time} hours
- Gaming Engagement Time: {gaming_time} hours

Predicted Stress Level: {predicted_stress}

Explain why the user's stress level is {predicted_stress} and provide personalized suggestions to reduce or maintain it.
"""

response = model_gemini.generate_content(prompt)

# ðŸ”¹ Display the AI-Generated Report
print("\nðŸ“Œ Stress Report:")
print(response.text)
```
